```python
# create object from the R using package Seurat and also calculate the mitoratio for the data: 
%%R
suppressPackageStartupMessages(library(Seurat))
library(dplyr)

setwd("/Users/macbook/Desktop/Bio-Research\ /PS050_cellranger_count_outs/raw_feature_bc_matrix")

EpCAM.data <- Read10X(data.dir = "/Users/macbook/Desktop/Bio-Research\ /PS050_cellranger_count_outs/filtered_feature_bc_matrix/")

# Create Seurat object
EpCAM <- CreateSeuratObject(counts = EpCAM.data, project = "mSG_EpCAM", save.SNN = TRUE)

# Add number of genes per UMI for each cell to metadata

EpCAM$log10GenesPerUMI <- log10(EpCAM$nFeature_RNA) / log10(EpCAM$nCount_RNA)

# Compute percent mito ratio
# What is the meaning of `pattern = "^mt-" ` here? 
EpCAM$mitoRatio <- PercentageFeatureSet(object = EpCAM, pattern = "^mt-")
EpCAM$mitoRatio <- EpCAM@meta.data$mitoRatio / 100

# Create metadata dataframe
metadata <- EpCAM@meta.data

# Add cell IDs to metadata
metadata$cells <- rownames(metadata)

# Rename columns
metadata <- metadata %>%
  dplyr::rename(seq_folder = orig.ident,
                nUMI = nCount_RNA,
                nGene = nFeature_RNA)

# Create sample column
metadata$sample <- "mSG"

# Create metadata dataframe
metadata <- EpCAM@meta.data

# Add cell IDs to metadata
metadata$cells <- rownames(metadata)

# Rename columns
metadata <- metadata %>%
  dplyr::rename(seq_folder = orig.ident,
                nUMI = nCount_RNA,
                nGene = nFeature_RNA)
# Create sample column
metadata$sample <- "mSG"

# Add metadata back to Seurat object
EpCAM@meta.data <- metadata

EpCAM
```

```python
# transform the R object into anndata in Python 

%%R -o EpCAM_sce2
#convert the Seurat object to a SingleCellExperiment object
EpCAM_sce2 <- as.SingleCellExperiment(EpCAM)

EpCAM_sce2
```

```python
# here, you could see that the EpCAM is a anndata object (with common used attributes like: X, obs, var_names 

# X, with 8277 observations and 32285 genes as features

# obs, pandas Data Frame, store the data nCount_RNA, nFeature_RNA, log10GenesPerUMI, mitRatio, and more information 

# var_names store all of the gene names as a list: 

# assign names to obs
EpCAM_sce2.obs.columns = ['seq_folder', 'nUMI', 'nGene', 'log10GenesPerUMI', 'mitoRatio', 'cells',
       'sample', 'ident']

# due to the nFeature_RNA is object, to make it consistence in some way, we need to transform it into float: 
EpCAM_sce2.obs.nUMI = EpCAM_sce2.obs.nUMI.astype(float)

# output basic information for EpCAM_sce2 (anndata type)
EpCAM_sce2
```
